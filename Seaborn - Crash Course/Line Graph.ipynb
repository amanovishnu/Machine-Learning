```python
import seaborn as sns;
import numpy as np;
import pandas as pd;
```

```python
sales = pd.read_csv('Datasets/sales.csv')
sns.relplot('Order Date','Sales',data=sales, kind='line');
```

```python
sales = pd.read_excel('Datasets/sales.xls',sheet_name=1)
sales['year'] = sales['Order Date'].dt.year
# CI = Confidence Interval
sns.relplot('year','Sales',data=sales, kind='line')
sales['year'].unique()
```

```python
sns.relplot('year','Sales',data=sales, kind='line', ci=None)
```

```python
sns.relplot('year','Sales',data=sales, kind='line', ci='sd')
```

```python
# Estimator Plots the Real Values
sns.relplot('year','Sales',data=sales, kind='line', estimator=None)
```

```python
sns.relplot('year','Sales',data=sales, kind='line', estimator=sum)
```

```python
sns.relplot('year','Sales',data=sales, kind='line', hue='Ship Mode')
```

```python
sns.relplot('year','Sales',data=sales, kind='line', style="Ship Mode", hue="Ship Mode")
```

```python
sns.relplot('year','Sales',data=sales, kind='line', col="Product Category", hue="Ship Mode")
```

```python
sns.relplot('year','Sales',data=sales, kind='line', col="Region", col_wrap=4, hue="Ship Mode")
```
