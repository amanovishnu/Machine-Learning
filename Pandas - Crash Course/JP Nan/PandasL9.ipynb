### How to Tackle null values in a dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Datasets/titanicKaggle.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.heatmap(df.isna())
```

```python
df[df['Age'].isna()]['Age']
```

```python
df['Age'].fillna(value ='good')
# df['Age'].fillna(value ='good',inplace=True) Use inplace to reflect in df
```

```python
df.isna().sum()
```

```python
df.fillna(method='backfill').head(10)
```

```python
df.fillna(method='bfill').head(10)
```

```python
df.fillna(method='ffill').head(10)
```

```python
df.Age.median()
```

```python
df['Age'].fillna(value = df['Age'].median() ,inplace=True)
```

```python
df.head(10)
```

```python
df.isna().sum()
```
